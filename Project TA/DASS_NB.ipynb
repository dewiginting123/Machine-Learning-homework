# Depresi

```python
def category_depresi(row):
   if row['score_D'] >= 0 and row['score_D'] <= 9:
    return "normal"
   if row['score_D'] >= 10 and row['score_D'] <= 13:
    return "ringan"
   if row['score_D'] >= 14 and row['score_D'] <= 20:
    return "sedang"
   if row['score_D'] >= 21 and row['score_D'] <= 27:
    return"parah"
   if row['score_D'] >= 28:
    return"sangat parah"
```
